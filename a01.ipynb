```python
# converting a proposition to its counter part inverse, converse and contrapositve

def convert_prop(string, prop):
    if prop == "implication":
        inverse = to_inverse(string)
        print("Inverse is : "+inverse)
        converse = to_converse(string)
        print("Converse is : "+converse)
        contrapositive = to_contrapositive(string)
        print("Contrapositive is : "+contrapositive)
        
    elif prop == "converse":
        converse = to_converse(string)
        print("Implication is : "+ converse)
        inverse = to_inverse(converse)
        print("Inverse is : "+ inverse)
        contrapositive = to_contrapositive(converse)
        print("Contrapositive is : "+contrapositive)
        
    elif prop == "inverse":
        inverse = to_inverse(string)
        print("Implication is : "+inverse)
        converse = to_converse(inverse)
        print("Converse is : "+converse)
        contrapositive = to_contrapositive(inverse)
        print("Contrapositive is : "+contrapositive)
    elif prop == "contrapositive":
        contrapositive = to_contrapositive(string)
        print("Implication is : "+ contrapositive)
        inverse = to_inverse(contrapositive)
        print("Inverse is : "+inverse)
        converse = to_converse(contrapositive)
        print("Converse is : "+converse)
    return
```

```python
def to_inverse(string):
    
    neg = {"-p":"p","-q":"q"}
    pos = {"p":"-p","q":"-q"}
    
    if string[0:2] in neg:
        
        
        string = string.replace(string[0:2],neg[string[0:2]])
        
        if string[len(string)-2:] in neg:
            string = string.replace(string[len(string)-2:],neg[string[len(string)-2:]])
        
        
        elif string[len(string)-1:] in pos:
            string = string.replace(string[len(string)-1:],pos[string[len(string)-1:]])
    
    
    elif string[0:1] in pos:
        
        
        string = string.replace(string[0:1],pos[string[0:1]])
        
        if string[len(string)-2:] in neg:
            string = string.replace(string[len(string)-2:],neg[string[len(string)-2:]])
        
        
        elif string[len(string)-1:] in pos:
            string = string.replace(string[len(string)-1:],pos[string[len(string)-1:]])
    
    
    
    
    return string
```

```python
def to_converse(string):
    
    neg = ["-p","-q"]
    pos = ["p","q"]
    
    if string[0:2] in neg:
        
        first = string[0:2]
        if string[len(string)-2:] in neg:
            second = string[len(string)-2:]
        elif string[len(string)-1:] in pos:
            second = string[len(string)-1:]
    
    elif string[0:1] in pos:
        
        first = string[0:1]
        if string[len(string)-2:] in neg:
            second = string[len(string)-2:]
        
        elif string[len(string)-1:] in pos:
            second = string[len(string)-1:]
    
    string = second +"->"+ first
    
    return string
```

```python
def to_contrapositive(string):
    string = to_inverse(string)
    string = to_converse(string)
    return string
```
